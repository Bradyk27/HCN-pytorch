```python
import numpy as np

print('example video 1 (subject 1)')
example_video = np.genfromtxt('lidar_mocap/video01subject01.csv', delimiter=',');

print('number of frames:',example_video.shape[0])
frame_number = 20;

print('example frame number:', frame_number)
frame = example_video[frame_number]

xs = frame[1::3];
ys = frame[2::3];
zs = frame[0::3];
zs = -1*zs;
zs = zs - np.amin(zs)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig1 = plt.figure()
ax1 = fig1.add_subplot(projection='3d')
#ax1.set_xlim3d(-5,5)
#ax1.set_ylim3d(50,60)
#ax1.set_zlim3d(0,10)
ax1.set_xlabel('lateral distance from lidar')
ax1.set_ylabel('depth distance from lidar')
ax1.set_zlabel('height from ground')
ax1.scatter(xs, ys, zs)
plt.show()

fig2 = plt.figure()
ax2 = fig2.add_subplot(projection='3d')
#ax2.set_xlim3d(-15,15)
#ax2.set_ylim3d(30,60)
#ax2.set_zlim3d(0,30)
ax2.set_xlabel('lateral distance from lidar (ft)')
ax2.set_ylabel('depth distance from lidar (ft)')
ax2.set_zlabel('height from ground (ft)')
ax2.scatter(xs, ys, zs)
plt.show()
```
